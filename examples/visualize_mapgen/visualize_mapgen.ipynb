# Make Cart Disc

```python
disc_pos = np.array([20, 20])
radius = 1.0
focus_pos = [70, 50]
num_points = 5
plot_disc = True
use_spiral = False
disc2 = make_cart_disc(disc_pos=disc_pos, radius=radius, focus_pos=np.asarray(focus_pos), num_points=num_points, plot_disc=False, use_spiral=use_spiral)

import matplotlib.pyplot as plt

from kwave.utils.data import scale_SI

print(np.shape(disc2))

_, scale, prefix, unit = scale_SI(np.max(disc2))

# create the figure
fig = plt.figure()

cmap = plt.get_cmap('viridis', np.shape(disc2)[1])
ax = fig.add_subplot(111, )
ax.scatter(disc2[0, :] * scale, disc2[1, :] * scale, marker='.',
          c=np.arange(np.shape(disc2)[1]), cmap=cmap, alpha=0.9, edgecolor=None)
ax.set_xlabel(f"[{prefix}m]")
ax.set_ylabel(f"[{prefix}m]")
ax.axis("equal")
ax.grid(True)
```

# Cart Spherical Segment

```python
bowl_pos = np.array([0.0, 0.0, 0.0])
focus_pos = np.array([0.0, 0.0, 1.0])
num_points_vec = [69, 70, 200, 1000]
radius = 65e-3
ap_diam1 = 30e-3
inner_diameter = 45e-3
outer_diameter = 60e-3

# find position where to split the bowl
bowl_height1 = radius - np.sqrt(radius**2 - (ap_diam1 / 2)**2)
bowl_height2 = radius - np.sqrt(radius**2 - (inner_diameter / 2)**2)

plot_bowl = True

# loop over some different numbers of points
for points_ind in np.arange(0, 1, dtype=int):

    num_points = num_points_vec[points_ind];

    # make regular bowl
    bowl_points = make_cart_bowl(bowl_pos, radius, outer_diameter, focus_pos, num_points)

    # split array
    bowl_points_ann1 = bowl_points[:, bowl_points[2, :] <= bowl_height1]
    bowl_points_ann2 = bowl_points[:, (bowl_points[2, :] <=  bowl_height2) & (bowl_points[2, :] > bowl_height1)]
    bowl_points_ann3 = bowl_points[:, bowl_points[2, :] > bowl_height2]

    # count points
    num_points_ann1 = np.shape(bowl_points_ann1)[1]
    num_points_ann2 = np.shape(bowl_points_ann2)[1]
    num_points_ann3 = np.shape(bowl_points_ann3)[1]

    num_points_inner = num_points_ann1 + num_points_ann2
    num_points = num_points_ann3

    make_cart_spherical_segment(bowl_pos, radius, inner_diameter, outer_diameter, focus_pos, num_points, plot_bowl, num_points_inner)

```
