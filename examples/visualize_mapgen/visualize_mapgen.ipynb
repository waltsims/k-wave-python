# Cart Rect

This takes the data from one of the test examples.

```python
rect_pos = [20, 20, 30]
Lx = 10
Ly = 10
theta = [70, 50, 20]
num_points = 500
data_3d = make_cart_rect(rect_pos, Lx, Ly, theta=theta, num_points=num_points, plot_rect=True)

rect_pos = [-20, -30]
Lx = 15
Ly = 30
theta = -5.3
num_points = 72
data_2d = make_cart_rect(rect_pos, Lx, Ly, theta=theta, num_points=num_points, plot_rect=True)
```

```python
from matplotlib import cm
fig = plt.figure()
cmap = plt.get_cmap('viridis', np.shape(data_3d)[1])
if len(rect_pos) == 3:
    ax = fig.add_subplot(111, projection="3d")
    ax.scatter(data_3d[0], data_3d[1], data_3d[2], marker='.', c=np.arange(np.shape(data_3d)[1]), cmap=cmap, alpha=0.9, edgecolor=None)
```
